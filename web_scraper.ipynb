```python
from bs4 import BeautifulSoup
import requests
import re 
import pandas as pd 
```

```python
url = 'http://www.imdb.com/chart/top'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
```

IMDB WEB SCRAPER 

```python
movies = soup.select('td.titleColumn')
crew = [a.attrs.get('title') for a in soup.select('td.titleColumn a')]
ratings = [b.attrs.get('data-value') for b in soup.select('td.posterColumn span[name=ir]')]
```

```python
list = []

for i in range(0, len(movies)):
  movie_string = movies[i].get_text()
  movie = (' '.join(movie_string.split()).replace('.',''))
  movie_title = movie[len(str(i))+1:-7]
  year = re.search('\((.*?)\)', movie_string).group(1)
  place = movie[:len(str(i))-(len(movie))]
  data = {"place": place,
          "movie_title": movie_title,
          "rating": ratings[i],
          "year": year,
          "star_cast": crew[i],
          }
  list.append(data)
```

```python
for movie in list:
    print(movie['place'], '-', movie['movie_title'], '('+movie['year'] +
        ') -', 'Starring:', movie['star_cast'], movie['rating'])
```

```python
df = pd.DataFrame(list)
df.to_csv('imdb_top_250_movies.csv',index=False)
```

```python
num_ratings = num_ratings.split()
num_ratings = num_ratings[0]
title = title.strip()
today = datetime.date.today()
```

```python
import csv
header = ['title', 'num_rating','date']
df = [title, num_ratings, today]
with open('web_screper2.csv','w', newline = '', encoding = 'UTF8') as f:
  writer = csv.writer(f)
  writer.writerow(header)
  writer.writerow(df)
```
